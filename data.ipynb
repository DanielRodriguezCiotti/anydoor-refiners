```python
from src.training.data.vitonhd import VitonHDDataset
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = VitonHDDataset(image_dir='dataset/test/cloth', filtering_file='dataset/lora_test_images.txt', inference=True)
```

```python
def display_images(images):
    fig, axs = plt.subplots(1, len(images), figsize=(15, 5))
    for i, (title, img) in enumerate(images.items()):
        # Rescale image if it's in the range [-1, 1] to [0, 1]
        if np.min(img) < 0 or np.max(img) > 1:
            img = (img + 1) / 2  # Rescale from [-1, 1] to [0, 1]
        axs[i].imshow(img)
        axs[i].axis("off")
        axs[i].set_title(title)
    plt.show()
```

```python
# Display the first 10 images
from tqdm import tqdm
imgs = []
for i in tqdm(range(100)):
    try:
        sample = dataset.get_sample(i)
        imgs.append(sample["jpg"])
        if len(imgs) == 5:
            display_images({
                f"Image {t}": imgs[t] for t in range(5)
            })
            imgs = []
    except Exception as e:
        print(e)
        continue
    
```
